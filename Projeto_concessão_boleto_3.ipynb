```python
def plot_bin(binx, title, show_iv):
    '''
    plot binning of one variable
    
    Params
    ------
    binx:
    title:
    show_iv:
    
    Returns
    ------
    matplotlib fig object
    '''
    # y_right_max
    y_right_max = np.ceil(binx['badprob'].max()*10)
    if y_right_max % 2 == 1: y_right_max=y_right_max+1
    if y_right_max - binx['badprob'].max()*10 <= 0.3: y_right_max = y_right_max+2
    y_right_max = y_right_max/10
    if y_right_max>1 or y_right_max<=0 or y_right_max is np.nan or y_right_max is None: y_right_max=1
    ## y_left_max
    y_left_max = np.ceil(binx['count_distr'].max()*10)/10
    if y_left_max>1 or y_left_max<=0 or y_left_max is np.nan or y_left_max is None: y_left_max=1
    # title
    title_string = binx.loc[0,'variable']+"  (iv:"+str(round(binx.loc[0,'total_iv'],4))+")" if show_iv else binx.loc[0,'variable']
    title_string = title+'-'+title_string if title is not None else title_string
    # param
    ind = np.arange(len(binx.index))    # the x locations for the groups
    width = 0.35       # the width of the bars: can also be len(x) sequence
    ###### plot ###### 
    fig, ax1 = plt.subplots()
    ax2 = ax1.twinx()
    # ax1
    p1 = ax1.bar(ind, binx['good_distr'], width, color=('#31F60A', '#31F60A', '#31F60A'))
    p2 = ax1.bar(ind, binx['bad_distr'], width, bottom=binx['good_distr'], color=('#31F60A', '#31F60A', '#31F60A'))
    for i in ind:
        ax1.text(i, binx.loc[i,'count_distr']*1.02, str(round(binx.loc[i,'count_distr']*100,1))+'%', ha='center')
    # ax2
    ax2.plot(ind, binx['badprob'], marker='o', color='black') # modifiquei aqui
    for i in ind:
        ax2.text(i, binx.loc[i,'badprob']+0.02, str(round(binx.loc[i,'badprob']*100,1))+'%', color='black', ha='center')
    # settings    
    ax1.set_ylabel('Bin count distribution')
    ax2.set_ylabel('Bad probability', color='black')
    ax1.set_yticks(np.arange(0, y_left_max+0.2, 0.2))
    ax2.set_yticks(np.arange(0, y_right_max+0.2, 0.1))
    ax2.tick_params(axis='y', colors='black')
    plt.xticks(ind, [s.replace('%','') for s in binx['bin']], rotation = '90')
    plt.title(title_string, loc='left')
    #plt.legend((p2[0], p1[0]), ('bad', 'good'), loc='upper right')
    #plt.legend((p2[0], p1[0]), ('distribution'), loc='upper right')
    plt.setp(ax1.xaxis.get_majorticklabels(), rotation=20 )
    # show plot 
    # plt.show()
    #plt.savefig(str(title_string) +'.png')
    #return print(title)
    return fig

def woebin_plot(bins, x=None, title=None, show_iv=True):
    '''
    WOE Binning Visualization
    ------
    `woebin_plot` create plots of count distribution and bad probability 
    for each bin. The binning informations are generates by `woebin`.
    
    Params
    ------
    bins: A list or data frame. Binning information generated by `woebin`.
    x: Name of x variables. Default is None. If x is None, then all 
      variables except y are counted as x variables.
    title: String added to the plot title. Default is None.
    show_iv: Logical. Default is True, which means show information value 
      in the plot title.
    
    Returns
    ------
    dict
        a dict of matplotlib figure objests
        
    Examples
    ------
    import scorecardpy as sc
    import matplotlib.pyplot as plt
    
    # load data
    dat = sc.germancredit()
    
    # Example I
    dt1 = dat[["creditability", "credit.amount"]]
    
    bins1 = sc.woebin(dt1, y="creditability")
    p1 = sc.woebin_plot(bins1)
    plt.show(p1)
    
    # Example II
    bins = sc.woebin(dat, y="creditability")
    plotlist = sc.woebin_plot(bins)
    
    # # save binning plot
    # for key,i in plotlist.items():
    #     plt.show(i)
    #     plt.savefig(str(key)+'.png')
    '''
    xs = x
    # bins concat 
    if isinstance(bins, dict):
        bins = pd.concat(bins, ignore_index=True)
    # good bad distr
    def gb_distr(binx):
        binx['good_distr'] = binx['good']/sum(binx['count'])
        binx['bad_distr'] = binx['bad']/sum(binx['count'])
        return binx
    bins = bins.groupby('variable').apply(gb_distr)
    # x variable names
    if xs is None: xs = bins['variable'].unique()
    # plot export
    plotlist = {}
    for i in xs:
        binx = bins[bins['variable'] == i].reset_index()
        plotlist[i] = plot_bin(binx, title, show_iv)
    return plotlist
```

```python
def dados_agrupados(df, bad_col, safra_col, bins_ajust):
    
    '''
    Função cria um dataframe com as variáveis agrupadas categóricas
    '''
    
    df_bin = pd.DataFrame()
    
    for coluna in list(df.select_dtypes(include = ['float64','int64']).columns):     
        if coluna != bad_col and coluna != safra_col:
            bins_temp = [float(i)-0.001 for i in list(bins_ajust[coluna].breaks) if i !='missing']
            bins_temp = [-np.inf] + bins_temp
            df_bin[coluna] = pd.cut(df[coluna], bins=bins_temp)
            df_bin[coluna] = df_bin[coluna].cat.add_categories('missing')
            df_bin.loc[df[coluna].isnull(), coluna] = 'missing'
    
    for coluna in list(df.select_dtypes(include = ['object','category']).columns):
        df_bin[coluna] = df[coluna]
        for x in range(0, len(bins_ajust[coluna].breaks)):
            df_bin.loc[df_bin[coluna].isin(bins_ajust[coluna].breaks[x].replace('%','').split(',')), coluna] = bins_ajust[coluna].breaks[x].replace('%','')
            df_bin.loc[df[coluna].isnull(), coluna] = 'missing'
        
    df_bin[safra_col] = pd.to_datetime(df[safra_col])
    df_bin[bad_col] = df[bad_col]
    
    return df_bin
```

```python
def ks_por_safra(df, bad_col, score_cols, safra_col):
    '''
    Função que calcula o KS por safra
    '''
    
    safra_list = sorted(list(df[safra_col].value_counts().index)) # lista com safras
    df_temp = pd.DataFrame() # Iniciando dataframe que vai armazenar os ks's
    
    for score_col in score_cols:
        ks_list = [] # lista com ks
    
        for safra in safra_list:
            ks_list.append(ks_2samp(df[(df[bad_col]==0) & (df[safra_col]==safra)][score_col],
                                    df[(df[bad_col]==1) & (df[safra_col]==safra)][score_col])[0])

        df_temp = pd.concat([df_temp, pd.DataFrame(ks_list,safra_list, columns=[score_col])], axis = 1)
    
    fig, ax = plt.subplots()
    df_temp.plot(ax=ax)
    plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
    plt.ylabel('KS')
    plt.xlim([dt.date(2019, 5, 1), dt.date(2020, 1, 1)])
    ax.axvspan(dt.date(2019, 6, 1), dt.date(2019, 8, 1), alpha=0.2, color='red')
    plt.ylim([0,0.35])
    
    return None
```

```python
def ks_por_safra_df(df, bad_col, score_cols, safra_col):
    '''
    Função que calcula o KS por safra
    '''
    
    safra_list = sorted(list(df[safra_col].value_counts().index)) # lista com safras
    df_temp = pd.DataFrame() # Iniciando dataframe que vai armazenar os ks's
    
    for score_col in score_cols:
        ks_list = [] # lista com ks
    
        for safra in safra_list:
            ks_list.append(ks_2samp(df[(df[bad_col]==0) & (df[safra_col]==safra)][score_col],
                                    df[(df[bad_col]==1) & (df[safra_col]==safra)][score_col])[0])

        df_temp = pd.concat([df_temp, pd.DataFrame(ks_list,safra_list, columns=[score_col])], axis = 1)
    
    
    return df_temp
```

```python
def default_por_aprovacao(df, bad_col, score_col, quantis):    

    bins = pd.qcut(df[score_col], q = quantis, duplicates='drop', retbins= True)[1] # obtendo bins
    df['col_bin'] = pd.qcut(df[score_col], q = quantis, labels = list(np.arange(1,len(bins))), duplicates='drop') # Fazendo a divisão dos scores
    df['col_bin'] = df['col_bin'].cat.add_categories('mis') # adicionando categoria missing
    df['col_bin'] =df['col_bin'].fillna('mis') # fill missing

    # Dataframe com volume
    df_vol = (pd.DataFrame(df['col_bin'].value_counts())/df[score_col].shape[0]).sort_index() # coluna de volumetria
    df_vol.rename(columns = {'col_bin': 'volume'}, inplace = True) # renomeando

    # Dataframe com bad ration (default)
    df_bad = df[~df[bad_col].isnull()][['col_bin',bad_col]].groupby(['col_bin']).mean()
    #df_bad['mau_abs'] = (df[~df[bad_col].isnull()][['col_bin',bad_col]].groupby(['col_bin']).sum())[bad_col]

    # Concat de volume e bad ration
    df_res = pd.concat([df_vol, df_bad], axis = 1)

    # volume acumulado
    df_res['volume_cum'] = df_res['volume'].cumsum()

    # volume x mau
    df_res['volume_mau'] = df_res['volume']*df_res[bad_col]

    # volume x mau (acumulado)
    df_res['volume_mau_cum'] = df_res['volume_mau'].cumsum()

    # bad ratio final
    df_res['mau_final'] = df_res['volume_mau_cum']/df_res['volume_cum']

    return df_res
```

```python
def criacao_variaveis_koin(df):
    
    '''
    Função que cria as colunas das novas variáveis provenientes do banco de dados da Koin
    '''
    
    #1dig_cep
    df['1dig_cep'] = df['json_cep'].str[:1]
    
    #2dig_cep
    df['2dig_cep'] = df['json_cep'].str[:2]
    
    #dist_origem_destino
    df['dist_origem_destino'] = np.nan # Inicializando
    for i in list(df[(~df['lat_origem'].isnull()) & (~df['lat_destino'].isnull())].index):
    
       df['dist_origem_destino'].loc[i] = geopy.distance.geodesic((df['lat_origem'][i],df['lon_origem'][i]), (df['lat_destino'][i],df['lon_destino'][i])).km

        
    #preco_por_distancia
    df.loc[df['dist_origem_destino'] > 0, 'preco_por_distancia'] = df['valor_pedido']/df['dist_origem_destino']
    
    #dominio_email
    df['dominio_email'] = df['json_email'].str.split('@').str[1].str.split('.').str[0]
    
    #tipo_complemento
    df.loc[df['json_complemento'].str.contains("casa", na=False, case=False), 'tipo_complemento'] = 'casa'
    df.loc[df['json_complemento'].str.contains("apart", na=False, case=False), 'tipo_complemento'] = 'apartamento'
    df.loc[df['json_complemento'].str.contains("apt ", na=False, case=False), 'tipo_complemento'] = 'apartamento'
    df.loc[df['json_complemento'].str.contains("ap ", na=False, case=False), 'tipo_complemento'] = 'apartamento'
    
    #idade
    df['idade'] = (pd.to_datetime(df['dt_create']) - pd.to_datetime(df['cad_dt_nascimento']))/ np.timedelta64(1, 'Y')
    
    #dias_ate_viagem
    df['dias_ate_viagem'] = (pd.to_datetime(df['delivery_date']) - pd.to_datetime(df['dt_create'])).dt.days
    
    #valor_pedido_diario_ate_viagem
    df['valor_pedido_diario_ate_viagem'] = df['valor_pedido']/df['dias_ate_viagem']
    
    #hora_do_pedido
    df['hora_do_pedido'] = pd.to_datetime(df['dt_create']).dt.hour
    
    #mes_viagem
    df['mes_viagem'] = pd.to_datetime(df['delivery_date']).dt.month
    
    #dia_da_semana_pedido
    df['dia_da_semana_pedido'] = pd.to_datetime(df['dt_create']).dt.weekday.replace({0:'monday',1:'tuesday',2:'wednessday',3:'thursday',4:'friday',5:'saturday',6:'sunday'})
    
    #dia_da_semana_pedido
    df['dia_da_semana_viagem'] = pd.to_datetime(df['delivery_date']).dt.weekday.replace({0:'monday',1:'tuesday',2:'wednessday',3:'thursday',4:'friday',5:'saturday',6:'sunday'})
    
    #qtd_dias_feriado_mais_prox
    df['qtd_dias_feriado_mais_prox'] = np.nan #Inicializando
    df['delivery_date'] = pd.to_datetime(df['delivery_date']) # Garantindo tipo certo da variável
    for i in list(df[~df['delivery_date'].isnull()].index):
    
       df['qtd_dias_feriado_mais_prox'].loc[i] = (df['delivery_date'].loc[i] - nearest(feriados['delivery_date'], df['delivery_date'].loc[i])).days    
    
    
    #dist_cidade_origem
    df['dist_cidade_origem'] = np.nan # Inicializando
    for i in list(df[(~df['lat_origem'].isnull()) & (~df['lat_cidade'].isnull())].index):
    
       df['dist_cidade_origem'].loc[i] = geopy.distance.geodesic((df['lat_origem'][i],df['lon_origem'][i]), (df['lat_cidade'][i],df['lon_cidade'][i])).km
    
    #rota
    df['rota'] = df['origem']+'-'+df['destino']
    
    #dia_semana_pedido_viagem
    df['dia_semana_pedido_viagem'] = df.apply(agrupa_dia_pedido_viagem, axis=1)

    
    return df

```
